```python
import pandas as pd
# Added placeholder columns for 'station_paris' and 'station_antwerp'
data = {'station_london': [10, 12, 15, 11, 13],
        'station_paris': [8, 10, 13, 9, 11],
        'station_antwerp': [15, 18, 20, 16, 19]}
air_quality = pd.DataFrame(data)
display(air_quality)
```

```python
air_quality["london_mg_per_cubic"]=air_quality["station_london"]*1.882
```

```python
air_quality["ratio_paris_antwerp"]=air_quality["station_paris"]/air_quality["station_antwerp"]
```
